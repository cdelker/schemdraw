```python
import schemdraw
from schemdraw.parsing import logicparse
#schemdraw.use('svg')
#schemdraw.settextmode('text')
```

```python
logicparse('(not a) and b or c', gateH=.5)  # gateH=0.5 makes them line up nicer when there's only one child
```

```python
logicparse('not ((w and x) or (y and z))', outlabel='$\overline{Q}$')
```

```python
logicparse('((a xor b) and (b or c) and (d or e)) or ((w and x) or (y and z))')
```

```python
logicparse(' (a  b) & (c  d)')  # Using logic symbols
```

```python
logicparse('(not a) or (b and c)', outlabel='X')
```
