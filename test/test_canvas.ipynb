```python
fig, ax = plt.subplots()
ax.plot([0,1,2,3,4,5],[0,1,0,-1,0,1], marker='o', ls='')
ax.set_aspect('equal')
with schemdraw.Drawing(canvas=ax) as d:
    elm.Resistor()
```

```python
svg = ET.Element('svg')
svg.set('xmlns', 'http://www.w3.org/2000/svg')
svg.set('xml:lang', 'en')
svg.set('height', f'100')
svg.set('width', f'300')
svg.set('viewBox', f'0 0 300 100')
circ = ET.SubElement(svg, 'circle')
circ.set('cx', '20')
circ.set('cy', '25')
circ.set('r', '25')
circ.set('fill', 'orange')

with schemdraw.Drawing(canvas=svg) as d:
    d.config(inches_per_unit=.5)
    d += elm.Resistor().label(r'1$\Omega$').at((1,-1))
```

```python
d = schemdraw.Drawing()
d += elm.Resistor()
d.draw(showframe=True, canvas='svg')
```

```python
d = schemdraw.Drawing()
d += elm.Resistor()
d.draw(showframe=True, canvas='matplotlib')
```
